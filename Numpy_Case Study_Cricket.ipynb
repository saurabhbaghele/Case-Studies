### Find mean,median,IQR for Sachin, Dravid and India

```python
sachin = cric_data1 [:, 0]
dravid = cric_data1 [:, 1]
india = cric_data1 [:, 2]
```

```python
def stats (col):
    print ("Mean", np.mean(col))
    print ("Median", np.median (col))
    print ("IQR", np.percentile (col, 75) - np.percentile (col, 25))
```

```python
stats (sachin)
```

```python
stats (dravid)
```

```python
stats (india)
```
