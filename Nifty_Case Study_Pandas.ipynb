# What fraction of day did the market close higher than the previous day's close

```python

nifty [1:]
```

```python
nifty [:-1]
```

```python
nifty_difference = nifty.values[1:] - nifty.values[:-1]
nifty_difference
```

```python
nifty_fraction = np.sum (nifty_difference > 0) / len(nifty)
nifty_fraction
```
